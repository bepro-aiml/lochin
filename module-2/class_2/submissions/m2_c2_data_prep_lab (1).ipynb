## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# Count missing values per column
missing = df.isnull().sum()
# Calculate what percentage of each column's values are missing
# Formula: (missing count / total rows) × 100, rounded to 2 decimal places
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
# Combine both results into a single summary table
missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]
```

```python
# Fill missing numerical values with median (robust to outliers)
# Select all columns that contain numbers (int or float)
numerical_cols = df.select_dtypes(include=[np.number]).columns
# Loop through each numerical column one by one
for col in numerical_cols:
  # Only process columns that actually have missing values
    if df[col].isnull().sum() > 0:
       # Calculate the median of the column
        # Median is used instead of average (mean) because it is not
        # affected by extreme outliers (e.g., a $1,000,000 sale skewing the average)
        median_val = df[col].median()
         # Replace all missing (NaN) values in this column with the median
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns
# Loop through each categorical column one by one
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
       # Calculate the mode = the most frequently appearing value in the column
        # .mode() returns a list, so [0] picks the top (most common) value
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify: no more missing values
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")
```

## Task 3: Remove Duplicates (pre-built)

```python
# Check for duplicates
# df.duplicated() marks every row that is an exact copy of a previous row as True
# .sum() counts how many such duplicate rows exist in total
n_duplicates = df.duplicated().sum()
# Print the result so we know if there's a problem
print(f"Duplicate rows found: {n_duplicates}")

# Remove duplicates
if n_duplicates > 0:
  # drop_duplicates() keeps the FIRST occurrence of each row
    # and removes all subsequent copies
    # The cleaned result is saved back into df, replacing the old version
    df = df.drop_duplicates()
    print(f"After removal: {df.shape[0]} rows remain")
else:
    print("No duplicates to remove.")
```

## Task 5: Derived Features (TODO)

Create customer-level summary features. These are the building blocks for customer segmentation (Activity 4).

You need to create:
- **total_spending**: Total sales per customer
- **order_frequency**: Number of orders per customer
- **avg_order_value**: Average sales amount per order per customer

Hint: Use `df.groupby('Customer ID')` (or whatever the customer ID column is named).

```python
# First, identify the right column names
print("All columns:")
for col in df.columns:
    print(f"  {col}")
```

```python
# TODO: Create total_spending per customer
# Hint: df.groupby('Customer ID')['Sales'].sum()
#
# Replace column names below with the actual names from your dataset.

# customer_spending = df.groupby('???')['???'].sum()
# customer_spending.name = 'total_spending'

# Your code here:
# ═════════════════════════════════════════════════════════
# STEP: Calculate Total Spending Per Customer
# ═════════════════════════════════════════════════════════

# WHY THIS MATTERS:
# Raw data has one row per ORDER, not per customer
# A single customer may have placed many orders over time
# Grouping by Customer ID adds up all their purchases into one total
# This lets us answer: "Who are our biggest spenders?"

# ─────────────────────────────────────────────
# PART 1: Group all orders by Customer ID
#         and sum up the Sales for each customer
# ─────────────────────────────────────────────

customer_spending = (
    df.groupby('Customer ID')  # Group rows that belong to the same customer
    ['Sales']                  # Focus only on the Sales column
    .sum()                     # Add up all sales for each customer
    .round(2)                  # Round to 2 decimal places (clean currency format)
)

# Give the resulting Series a clear name for readability
customer_spending.name = 'total_spending'

# ─────────────────────────────────────────────
# PART 2: Sort customers from highest to lowest spender
# ─────────────────────────────────────────────

customer_spending = customer_spending.sort_values(ascending=False)

# ─────────────────────────────────────────────
# PART 3: Preview the results
# ─────────────────────────────────────────────

print(f"Total unique customers: {len(customer_spending)}")
print(f"\nTop 5 highest spending customers:")
print(customer_spending.head())

print(f"\nBottom 5 lowest spending customers:")
print(customer_spending.tail())
```

```python
# TODO: Create order_frequency per customer
# Hint: Count the number of rows (orders) per customer.
# Use .groupby(...).size() or .groupby(...)['some_col'].count()

# Your code here:


# ═════════════════════════════════════════════════════════
# STEP: Calculate Order Frequency Per Customer
# ═════════════════════════════════════════════════════════

# WHY THIS MATTERS:
# Total spending alone doesn't tell the full story
# A customer could spend $10,000 in 1 big order vs 100 small orders
# Order frequency reveals HOW OFTEN a customer buys, which helps with:
#   - Identifying loyal, repeat customers
#   - Spotting one-time buyers who never returned
#   - Building customer retention strategies

# ─────────────────────────────────────────────
# PART 1: Count number of orders per customer
# ─────────────────────────────────────────────

order_frequency = (
    df.groupby('Customer ID')  # Group all rows belonging to the same customer
    .size()                    # Count how many rows (orders) each customer has
    .sort_values(ascending=False)  # Rank from most frequent to least frequent buyer
)

# Give the resulting Series a clear, descriptive name
order_frequency.name = 'order_frequency'

# ─────────────────────────────────────────────
# PART 2: Preview the results
# ─────────────────────────────────────────────

print(f"Total unique customers: {len(order_frequency)}")

print(f"\nTop 5 most frequent buyers:")
print(order_frequency.head())

print(f"\nTop 5 least frequent buyers:")
print(order_frequency.tail())

# ─────────────────────────────────────────────
# PART 3: Summary statistics
# Understand the spread of buying behaviour across all customers
# ─────────────────────────────────────────────

print(f"\nOrder frequency summary:")
print(f"  Average orders per customer : {order_frequency.mean():.1f}")
print(f"  Highest order count         : {order_frequency.max()}")
print(f"  Lowest order count          : {order_frequency.min()}")
```

```python
# TODO: Create avg_order_value per customer
# Hint: Use .groupby(...)['Sales'].mean()

# Your code here:
# ═════════════════════════════════════════════════════════
# STEP: Calculate Average Order Value Per Customer
# ═════════════════════════════════════════════════════════

# WHY THIS MATTERS:
# Total spending tells us WHO spends the most overall
# Order frequency tells us WHO buys the most often
# Average order value (AOV) tells us HOW MUCH a customer spends per single order
# Together all three give a complete picture of customer buying behaviour:
#   - High spender + low frequency = big bulk buyer
#   - Low spender + high frequency = loyal small buyer
#   - High AOV + high frequency    = VIP customer

# ─────────────────────────────────────────────
# PART 1: Calculate average sale value per customer
# ─────────────────────────────────────────────

avg_order_value = (
    df.groupby('Customer ID')  # Group all rows belonging to the same customer
    ['Sales']                  # Focus only on the Sales column
    .mean()                    # Calculate the average sale value per customer
    .round(2)                  # Round to 2 decimal places (clean currency format)
    .sort_values(ascending=False)  # Rank from highest to lowest average spender
)

# Give the resulting Series a clear, descriptive name
avg_order_value.name = 'avg_order_value'

# ─────────────────────────────────────────────
# PART 2: Preview the results
# ─────────────────────────────────────────────

print(f"Total unique customers: {len(avg_order_value)}")

print(f"\nTop 5 customers by average order value:")
print(avg_order_value.head())

print(f"\nBottom 5 customers by average order value:")
print(avg_order_value.tail())

# ─────────────────────────────────────────────
# PART 3: Summary statistics
# Understand the spread of average order values across all customers
# ─────────────────────────────────────────────

print(f"\nAverage order value summary:")
print(f"  Overall average order value : ${avg_order_value.mean():.2f}")
print(f"  Highest average order value : ${avg_order_value.max():.2f}")
print(f"  Lowest average order value  : ${avg_order_value.min():.2f}")
```

```python
# TODO: Combine all three into a single customer-level DataFrame
# Hint: Use pd.concat([series1, series2, series3], axis=1)
# or create them all at once with .groupby(...).agg(...)

# customer_summary = pd.concat([customer_spending, order_freq, avg_order], axis=1)
# customer_summary.columns = ['total_spending', 'order_frequency', 'avg_order_value']

# Your code here:
# ═════════════════════════════════════════════════════════
# STEP: Combine All 3 Metrics Into One Customer Summary Table
# ═════════════════════════════════════════════════════════

# WHY THIS MATTERS:
# Right now the 3 metrics exist as separate Series objects
# Combining them into one DataFrame makes it easy to:
#   - View all metrics side by side per customer
#   - Sort and filter customers by multiple criteria at once
#   - Feed the data into clustering or segmentation models later

# ─────────────────────────────────────────────
# OPTION 1: Combine the 3 already-created Series (quick approach)
# Use this if you already ran the previous 3 steps above
# ─────────────────────────────────────────────

customer_summary = pd.concat(
    [customer_spending,   # Total sales per customer
     order_frequency,     # Number of orders per customer
     avg_order_value],    # Average sale value per customer
    axis=1                # axis=1 means stack them side by side as columns
                          # (axis=0 would stack them as rows — not what we want)
)

# Rename columns to clean, readable names
customer_summary.columns = ['total_spending', 'order_frequency', 'avg_order_value']

# ─────────────────────────────────────────────
# OPTION 2: Build everything in one single step using .agg() (cleaner approach)
# Use this if you want to skip creating the 3 separate Series above
# ─────────────────────────────────────────────

customer_summary = (
    df.groupby('Customer ID')  # Group rows by customer
    .agg(
        total_spending  = ('Sales', 'sum'),   # Sum of all sales per customer
        order_frequency = ('Sales', 'count'), # Number of orders per customer
        avg_order_value = ('Sales', 'mean')   # Average order value per customer
    )
    .round(2)                  # Round all values to 2 decimal places
    .sort_values('total_spending', ascending=False)  # Sort by biggest spender first
    .reset_index()             # Turn Customer ID from index into a regular column
)

# ─────────────────────────────────────────────
# PART 2: Preview and validate the combined table
# ─────────────────────────────────────────────

print(f"Customer summary shape: {customer_summary.shape}")  # Should be (n_customers, 4)
print(f"\nFirst 5 rows:")
print(customer_summary.head())

# ─────────────────────────────────────────────
# PART 3: Overall summary statistics
# Get a quick feel for the distribution of all 3 metrics at once
# ─────────────────────────────────────────────

print(f"\nSummary statistics:")
print(customer_summary.describe().round(2))
```

## Reflection

Answer these in a text cell or comments:

1. Why did we use median instead of mean for filling numerical missing values?
2. What is the difference between the row-level DataFrame (one row per order) and the customer-level summary? When would you use each?
3. If two rows have identical values in every column, are they always true duplicates? When might they not be?

1.The mean (average) is easily pulled toward extreme values (outliers). In retail data like Superstore, extreme values are very common — one corporate client might place a $50,000 order while most orders are under $500. If you fill missing values with the mean, you are injecting an unrealistically high number into your data, which distorts future analysis, averages, and charts.For example, imagine 5 orders with values of $10, $12, $11, $13, and $500. The mean comes out to $109.20, which is distorted by that one large order. The median however comes out to $12, which reflects the typical order much more honestly.The median is robust because it simply picks the middle value and is completely unaffected by how extreme the outliers are. That is why it is the safer and more honest choice for filling numerical gaps.
2.The row-level DataFrame has one row per order and contains every detail such as product name, order date, region, discount, and shipping mode. The customer-level summary has one row per customer and contains only aggregated metrics like total spending, order frequency, and average order value.
You use the row-level DataFrame when you need detail — for example, which products sold the most in Q3, what the average discount was by region, or how long shipping took per order.
You use the customer-level summary when you need the big picture — for example, who are the top 10 biggest spenders, which customers have only ordered once, or how to segment customers into VIP vs at-risk groups.
A simple way to think about it is that the row-level data is the full raw story, while the customer-level summary is the condensed chapter.
3.Not always. This is a genuinely tricky problem in real data.
Identical rows might be completely legitimate when a customer genuinely orders the same product twice but the date column was missing or corrupted, making two real orders look identical. They can also appear when a dataset was exported twice from a reporting tool, creating artificial copies of real distinct events. Sometimes two completely different customers in different regions happen to order the exact same product, quantity, and discount on the same day purely by coincidence.
The safe habit is to always ask whether the dataset has a unique identifier column like Order ID. If it does, use that to detect duplicates instead of comparing every column. That way you catch true duplicates where the same Order ID appears twice, without accidentally removing legitimate orders that just happen to look identical.
The bottom line is that drop_duplicates() is a blunt tool. In real data work, always verify with a unique key column before removing anything.
