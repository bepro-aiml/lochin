````python
# Import libraries for working with data, math operations, and graphs
# pandas is used for tables/dataframes
# numpy is used for numerical operations
# matplotlib is used for plotting charts
# warnings.filterwarnings ignores warning messages to keep output clean

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')

# Load the Telco Customer Churn dataset from GitHub into a dataframe
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)

# Convert TotalCharges column into numeric values
# If there are invalid values, replace them with 0
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)

# Print the number of rows and columns in the dataset
print('Loaded:', df.shape)
````

### Cell 1 — see the first rows

````python
# Display the first 5 rows of the dataframe
# head() helps quickly preview the dataset structure and values
df.head()
````

### Cell 2 — dimensions of the data

````python
# Show the size of the dataframe
# shape returns the number of rows and columns
df.shape
````

### Cell 3 — column names and types

````python
# Display all column names and their data types
# dtypes helps understand what kind of data each column contains
df.dtypes
````

### Cell 4 — count customers who churned vs stayed

````python
# Count how many customers stayed and how many churned
# value_counts() counts occurrences of each unique value
df['Churn'].value_counts()
````

### Cell 5 — only customers with tenure > 50 months

````python
# Filter customers who stayed longer than 50 months
# Create a new dataframe called long_term with only long-term customers
long_term = df[df['tenure'] > 50]

# Print the number of long-term customers
print('Long-term customers:', len(long_term))

# Show the first rows of the filtered dataframe
long_term.head()
````

### Cell 6 — top 5 customers by TotalCharges

````python
# Sort customers by TotalCharges in descending order
# head(5) returns the top 5 customers with the highest charges
top5 = df.sort_values('TotalCharges', ascending=False).head(5)

# Display selected columns for the top 5 customers
top5[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']]
````

### Cell 7 — average MonthlyCharges per contract type

````python
# Group customers by contract type
# Calculate the average MonthlyCharges for each contract category
# round(2) rounds results to 2 decimal places
df.groupby('Contract')['MonthlyCharges'].mean().round(2)
````

### Cell 8 — average MonthlyCharges AND tenure per Churn group

````python
# Group data by churn status (Yes or No)
# Calculate average MonthlyCharges and average tenure for each group
# agg() allows multiple calculations on different columns
df.groupby('Churn').agg({'MonthlyCharges': 'mean', 'tenure': 'mean'}).round(2)
````

### Cell 9 — count of senior citizens by Churn

````python
# Create a cross table between SeniorCitizen and Churn columns
# crosstab() shows how many senior and non-senior customers churned or stayed
pd.crosstab(df['SeniorCitizen'], df['Churn'])
````
