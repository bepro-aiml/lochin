```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/refs/heads/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
```

```python
counts = df['Contract'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(counts.index, counts.values, color='teal')
plt.title("Customers per Contract type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['tenure'], bins=30, color='coral', edgecolor='black')
plt.title('Distribution of Tenure')
plt.ylabel('Number of Customers')
plt.xlabel('Tenure (Months)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['MonthlyCharges'], df['TotalCharges'], alpha=0.3, color='navy')
plt.title('Monthly Charges vs. Total Charges')
plt.ylabel('Total Charges')
plt.xlabel('Monthly Charges')
plt.show()
```

```python
groups = [df[df['Contract'] == c] ['MonthlyCharges'] for c in df['Contract'].unique()]
labels = df['Contract'].unique()
plt.figure(figsize=(10, 6))
plt.boxplot(groups, labels=labels)
plt.title('Monthly Charges by Contract')
plt.ylabel('Montly Charges')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10, 6))
ax[0].bar(df['Churn'].value_counts().index, df['Churn'].value_counts().values, color='gray')
ax[0].set_title('Churn')
ax[1].hist(df['MonthlyCharges'], bins=30, color='orange')
ax[1].set_title('Monthly Charges')
ax[2].scatter(df['tenure'], df['MonthlyCharges'], alpha=0.2, s=5)
ax[2].set_title('Tenure vs. Monthly Charges')
plt.tight_layout()
plt.show()
```
