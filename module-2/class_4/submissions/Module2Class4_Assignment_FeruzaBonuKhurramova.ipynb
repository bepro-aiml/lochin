# Task 1.1

```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({
    "name": ["Ali", "Zara", "Jasur"],
    "age": [22, 28, 31],
    "city": ["Tashkent", "Samarkand", "Bukhara"]
})

# 1. Check df["age"] return type
print(f"Type of df['age']: {type(df['age'])}")

# 2. Check df[['name', 'age']] return type
print(f"Type of df[['name', 'age']]: {type(df[['name', 'age']])}")

# 3. Print the index
print("Default Index:", df.index)
```

What does df["age"] return? It returns a Series (a 1D labeled array).

What does df[["name", "age"]] return? It returns a DataFrame (a 2D table). Double brackets always return a DataFrame.

What is the index by default? It is a RangeIndex starting at 0 and ending at $n-1$ (in this case, start=0, stop=3, step=1).

# Task 3.2 — Boolean Filtering 

```python
# Filter: Passengers older than 30 AND were in First Class (Pclass == 1)
older_first_class = df[(df["Age"] > 30) & (df["Pclass"] == 1)]

# Filter: Passengers who paid a very high fare OR were in Third Class
high_fare_or_third = df[(df["Fare"] > 100) | (df["Pclass"] == 3)]

print(f"Count of older first class: {len(older_first_class)}")
```

# Why must we use & instead of and, and why are parentheses required?
We use & because Pandas performs "bitwise" operations across entire arrays of data at once, whereas the Python word and only works on single True/False values. Parentheses are required because of "operator precedence"—without them, Python might try to compare the middle values before evaluating the left and right conditions.

# Task 3.3 — GroupBy

```python
# Pattern 1: Simple mean of Age grouped by Sex
mean_age_by_sex = df.groupby("Sex")["Age"].mean()
print(mean_age_by_sex)

# Pattern 2: Multiple aggregations
grouped_stats = df.groupby("Pclass").agg({
    "Fare": "mean",
    "Survived": "count"
})
print(grouped_stats)
```

The first call (df.groupby("Sex")["Age"].mean()) returns a Series.

# Task 3.4 — Merging

```python
# Create the lookup table
class_labels = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Name": ["Elite", "Middle", "Economy"]
})

# Merge it into the main df
df = pd.merge(df, class_labels, on="Pclass", how="left")

# View the new column at the end
print(df[["Name", "Pclass", "Class_Name"]].head())
```

# What happens to rows in the left table that have no match in the lookup?
If a row in the left table (our Titanic data) has no matching key in the lookup table, the new columns will be filled with NaN (Not a Number) for that row. The original row is kept, but the "joined" data remains empty.
