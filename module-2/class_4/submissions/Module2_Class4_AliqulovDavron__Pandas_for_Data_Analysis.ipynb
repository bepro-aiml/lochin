# Filtering, GroupBy, Merging (maps to lecture section “Filtering, GroupBy, Merging”)


## Boolean filtering

```python
tips_df[(tips_df['total_bill'] >=20) & (tips_df['sex'] == 'Male')] # Short command and fast read  use &, parentheses not to avoid confusion
```

## GroupBy: split-apply-combine


```python
tips_df.groupby('sex')['total_bill'].mean() #Group gender men total bill
```

```python
tips_df.groupby('sex')['smoker'].count() # Gender smoker count
```

```python
tips_df.groupby('sex').agg({'total_bill': ['mean', 'min', 'max'], 'tip': ['mean', 'min','max'], 'size': ['mean', 'min', 'max'], 'smoker': 'count'}) # Describe gender elements
```

##Merging

```python
day_type_lookup = pd.DataFrame({
    'day': ['Thur', 'Fri', 'Sat', 'Sun'],
    'day_type': ['Weekday', 'Weekday', 'Weekend', 'Weekend']
})

print("Lookup DataFrame:")
display(day_type_lookup)
```

```python
tips_df_merged = pd.merge(tips_df, day_type_lookup, on='day', how='left')

print("Merged DataFrame (first 5 rows with new 'day_type' column):")
display(tips_df_merged.head())
```

```python
region_names = ['west', 'south', 'north', 'east']

# Create a Series of regions with the same length as tips_df
# This will be used to add a 'region' column to tips_df
tips_df_regions = pd.Series(np.random.choice(region_names, size=len(tips_df)))
```

```python
# Create a copy of tips_df to add the new 'region' column
tips_df_with_regions = tips_df.copy()
tips_df_with_regions['region'] = tips_df_regions

print("DataFrame with 'region' column (first 5 rows):")
display(tips_df_with_regions.head())
```
