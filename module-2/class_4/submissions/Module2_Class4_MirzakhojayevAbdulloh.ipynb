```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/refs/heads/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
df.shape
```

```python
df.info()
```

```python
df.describe().round(2)
```

```python
high_pay = df[df['MonthlyCharges'] > 80]
print(len(high_pay))
high_pay.head()
```

```python
top10 = df.sort_values('tenure', ascending=False).head(10)
top10
```

```python
df.groupby('PaymentMethod')['MonthlyCharges'].mean().round(2)
```

```python
df.groupby('Churn').agg({'TotalCharges': ['mean', 'count']}).round(2)
```

```python
pd.crosstab(df['Contract'], df['Churn'])
```
