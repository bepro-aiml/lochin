### Task 3.1 — Advanced Filtering ###

```python
# Task 3.1: Filtering by Multiple Conditions
# We want high-scoring modern games
modern_high_score = df[(df["pts"] > 150) & (df["year_id"] > 2010)]

print(f"Found {len(modern_high_score)} modern games with > 150 points.")
display(modern_high_score[["date_game", "team_id", "pts", "opp_id", "opp_pts"]])
```

### Task 3.2 — The Power of GroupBy ###

```python
# Task 3.2: Grouping and Aggregating
# Calculate average points per franchise and sort them
top_scoring_teams = df.groupby("fran_id")["pts"].mean().sort_values(ascending=False)

print("Top 5 Highest Scoring Franchises (Historical Average):")
print(top_scoring_teams.head(5))
```

### Section 4: Data Cleaning (Final Step for Part A) ###

```python
# Task 4.1: Handling Missing Values
# Since 'notes' is mostly empty, we will fill missing ones with 'no notes'
df["notes"] = df["notes"].fillna("no notes")

# Verify that there are now 0 missing values
print("Remaining missing values in 'notes':", df["notes"].isna().sum())
```

##### 2. Quick Check Answers #####
Why use & instead of and? We use & because it is a bitwise operator that can compare entire Pandas Series element-by-element. The standard and only works on single True/False values.

Why are parentheses required? Parentheses are mandatory around each condition because bitwise operators like & have higher precedence than comparison operators (like >). Without them, Python gets confused about the order of operations.

The Final Missing Piece: Task 3.4 — Merging

```python
# Task 3.4: Merging with a Lookup Table
# 1. Create a small lookup DataFrame manually
# This maps the Franchise ID to their Home City and Arena Capacity
lookup_df = pd.DataFrame({
    "fran_id": ["Condors", "Squires", "Floridians", "Stars"],
    "home_city": ["Pittsburgh", "Virginia", "Miami", "Utah"],
    "arena_capacity": [12500, 10000, 15000, 19000]
})

# 2. Merge it into your main NBA DataFrame
# We use a 'left' join to keep all our NBA games
df_final = pd.merge(df, lookup_df, on="fran_id", how="left")

# 3. Check the new columns at the end
print("New columns added via merge:")
display(df_final[["date_game", "fran_id", "home_city", "arena_capacity"]].head())
```

What happens to rows with no match? Since we used a left join, all original NBA rows are kept. For franchises not in our small lookup_df (like the Lakers or Celtics), the new home_city and arena_capacity columns will simply contain NaN (Not a Number) values.
