```python
products = ["CeraVe Foaming Facial Cleanser", "The Ordinary Squalane Cleanser", "Cetaphil Daily Facial Cleanser", "CeraVe Moisturizing Lotion", "Neutrogena Hydro Boost Gel", "Nivea Soft Cream", "The Ordinary Niacinamide 10%", "The Ordinary Alpha Arbutin 2%", "The Ordinary Hyaluronic Acid 2%", "CeraVe AM SPF 30", "Neutrogena Sunscreen", "Eucerin Sunscreen"]
product_prices = [285000, 110000, 310000, 200000, 210000, 98000, 285000, 250000, 120000, 260000, 240000, 330000]

print("--- Product Classification Report ---")

for i, price in enumerate(product_prices):
    product_name = products[i]
    if price > 200000:
        category = "Expensive/Premium"
    elif price >= 100000:
        category = "Standard Product"
    else:
        category = "Budget-friendly"
        
    print(f"Product {i+1}: {product_name} | Price: {price:,} UZS | Category: {category}")

```

```python
import matplotlib.pyplot as plt

# 1. calculating categories
counts = {"Expensive/Premium": 0, "Standard Product": 0, "Budget-friendly": 0}

for price in product_prices:
    if price > 200000:
        counts["Expensive/Premium"] += 1
    elif price >= 100000:
        counts["Standard Product"] += 1
    else:
        counts["Budget-friendly"] += 1

# 2. graphicing
names = list(counts.keys())
values = list(counts.values())

plt.bar(names, values, color=['red', 'blue', 'green'])
plt.ylabel('Number of Products')
plt.title('Product Distribution by Category')
plt.show()
```
