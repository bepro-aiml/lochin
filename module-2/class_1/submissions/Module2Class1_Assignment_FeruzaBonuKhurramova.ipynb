```python
def calculate_average(numbers):
    total = sum(numbers)    # Push this in with Tab
    count = len(numbers)    # Push this in with Tab
    return total / count    # Push this in with Tab

# These lines stay on the left because they are NOT part of the function
test1 = [80, 90, 70, 85, 95]
test2 = [55, 62, 48, 71]
print("Average 1:", calculate_average(test1))
print("Average 2:", calculate_average(test2))
```
