```python
# Function 1: Bills Stats
def describe_bills(bills):
    avg = sum(bills) / len(bills)
    minimum = min(bills)
    maximum = max(bills)
    return avg, minimum, maximum

# Function 2: Customer Classification
def classify_customer(avg_spend):
    if avg_spend > 150000:
        return "Premium"
    elif avg_spend >= 50000:
        return "Standard"
    else:
        return "Basic"

# Function 3: Churn Rate 
def calculate_churn_rate(total_customers, churned_customers):
    rate = (churned_customers / total_customers) * 100
    return rate

# --- EXECUTION ---
monthly_bills = [120000, 150000, 90000, 110000, 60000, 160000, 40000]
total_clients = 10000
lost_clients = 350

# Logic Flow
average, low, high = describe_bills(monthly_bills)
tier = classify_customer(average)
churn = calculate_churn_rate(total_clients, lost_clients)

# Final Output
print(f"--- Business Report ---")
print(f"Average Revenue per User: {average:,.0f} UZS")
print(f"Customer Segment: {tier}")
print(f"Monthly Churn Rate: {churn}%")
```
