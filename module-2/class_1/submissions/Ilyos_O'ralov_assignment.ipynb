```python
# Task 2: Create variables of 5 different types and print their values with types
age = 22
gpa = 3.85
name = "Aziz"
is_student = True
scores = [85, 92, 78, 95, 88]

print(age, type(age))
print(gpa, type(gpa))
print(name, type(name))
print(is_student, type(is_student))
print(scores, type(scores))
```

```python
# Task 3: Define calculate_average function and test it with two different lists
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

test_list_1 = [80, 90, 70, 85, 95]
test_list_2 = [55, 62, 48, 71]

print("Average 1:", calculate_average(test_list_1))
print("Average 2:", calculate_average(test_list_2))
```

```python
# Task 4: Iterate through a list of 5+ items and print each with its index
cities = ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana"]

for index, city in enumerate(cities):
    print(f"Index {index}: {city}")
```

```python
# Task 5: Generate random scores, calculate average, and check pass/fail status
import random

# Generate 10 random exam scores between 0 and 100
exam_scores = [random.randint(0, 100) for _ in range(10)]
print("Scores:", exam_scores)

# Calculate class average using the function from Task 3
class_avg = calculate_average(exam_scores)
print(f"Class average: {class_avg:.1f}")

# Determine pass/fail based on threshold of 60
if class_avg >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
```
