```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
```

```python
obj_cols = df.select_dtypes(include='object').columns.to_list()
print(f"Total: {len(obj_cols)}")
print(obj_cols)
```

```python
df['gender'].value_counts()
```

```python
dummies = pd.get_dummies(df['InternetService'], prefix='Internet')
print(dummies.shape)
print(dummies.head())
```

```python
dummies_2 = pd.get_dummies(df['InternetService'], prefix='Internet', drop_first=True)
print(dummies_2.shape)
print(dummies_2.head())
```

```python
encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df[['TechSupport']])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['TechSupport']))
encoded_df.head()
```

```python
df['Partner_bin'] = df['Partner'].map({'No': 0, "Yes": 1})
df[['Partner', 'Partner_bin']].head()
```

```python
multi = pd.get_dummies(df[['gender', 'InternetService', 'TechSupport', 'Contract']], drop_first=True)
print(multi.shape)
multi.head()
```
