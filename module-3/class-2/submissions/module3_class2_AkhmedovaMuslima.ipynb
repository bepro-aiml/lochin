````python
# Import libraries for data analysis, scaling, and plotting
# Load the Telco Customer Churn dataset into a DataFrame
# Convert TotalCharges to numeric values and replace invalid values with 0

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')

url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
print('Loaded:', df.shape)
````

### Cell 1 — see how different the ranges are

````python
# Create a list of numeric columns that will be used for scaling
cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

# Show summary statistics for the selected columns
df[cols].describe().round(2)
````

### Cell 2 — apply StandardScaler (mean=0, std=1)

````python
# Import StandardScaler for standardization
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Select the columns that will be scaled
X = df[cols]

# Learn mean and standard deviation, then scale the data
X_std = scaler.fit_transform(X)

# Convert the scaled array back into a DataFrame
X_std_df = pd.DataFrame(X_std, columns=cols)

# Display statistics of the scaled data
X_std_df.describe().round(2)
````

### Cell 3 — apply MinMaxScaler (range 0 to 1)

````python
# Import MinMaxScaler for scaling data between 0 and 1
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
mm = MinMaxScaler()

# Scale all values to the range from 0 to 1
X_mm = mm.fit_transform(X)

# Convert the scaled array into a DataFrame
X_mm_df = pd.DataFrame(X_mm, columns=cols)

# Show minimum and maximum values after scaling
X_mm_df.agg(['min', 'max']).round(2)
````

### Cell 4 — apply RobustScaler (uses median + IQR, ignores outliers)

````python
# Import RobustScaler which is resistant to outliers
from sklearn.preprocessing import RobustScaler

# Create a RobustScaler object
rb = RobustScaler()

# Scale the data using median and IQR
X_rb = rb.fit_transform(X)

# Convert the scaled array into a DataFrame
X_rb_df = pd.DataFrame(X_rb, columns=cols)

# Display summary statistics of the scaled data
X_rb_df.describe().round(2)
````

### Cell 5 — split FIRST, then scale (the big rule — no data leakage)

````python
# Import the train_test_split function
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

# Create a new StandardScaler object
scaler = StandardScaler()

# Fit the scaler on training data and transform it
X_train_s = scaler.fit_transform(X_train)

# Transform test data using the same scaler parameters
X_test_s = scaler.transform(X_test)

# Print the shapes of the train and test datasets
print(f'Train shape: {X_train_s.shape}, Test shape: {X_test_s.shape}')

# Check that the scaled training data has mean close to zero
print(f'Train mean (should be ~0): {X_train_s.mean(axis=0).round(2)}')
````

### Cell 6 — save the trained scaler to disk

````python
# Import joblib for saving and loading Python objects
import joblib

# Save the trained scaler to a file
joblib.dump(scaler, 'my_scaler.joblib')

# Load the saved scaler from the file
loaded = joblib.load('my_scaler.joblib')

# Print the mean values learned by the scaler
print('Saved and loaded. Learned mean:', loaded.mean_.round(2))
````
