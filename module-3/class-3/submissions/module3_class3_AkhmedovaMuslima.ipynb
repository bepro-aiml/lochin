````python
# Import the libraries needed for data analysis and visualization
# Load the customer churn dataset and convert TotalCharges to numeric values
# Replace invalid values with 0 so the column can be used correctly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')

url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
print('Loaded:', df.shape)
````

### Cell 1 — see which columns are categorical (text)

````python
# Find all columns that contain text (categorical data)
# This helps us understand which columns may need encoding before modeling

cat_cols = df.select_dtypes(include='object').columns.tolist()
print('Categorical columns:', cat_cols)
print('Total:', len(cat_cols))
````

### Cell 2 — look at one categorical column

````python
# Count how many customers belong to each contract type
# value_counts() shows the frequency of each category

df['Contract'].value_counts()
````

### Cell 3 — use pd.get_dummies to one-hot encode Contract

````python
# Convert the Contract column into separate binary columns using one-hot encoding
# Each category gets its own 0/1 column

dummies = pd.get_dummies(df['Contract'], prefix='Contract')
print('Shape before:', df[['Contract']].shape, '-> after:', dummies.shape)
dummies.head()
````

### Cell 4 — drop_first=True (to avoid the dummy variable trap)

````python
# Apply one-hot encoding again but remove the first category
# drop_first=True helps avoid duplicate information between encoded columns

dummies2 = pd.get_dummies(df['Contract'], prefix='Contract', drop_first=True)
print('With drop_first:', dummies2.shape)
dummies2.head()
````

### Cell 5 — use OneHotEncoder from sklearn

````python
# Use sklearn's OneHotEncoder to transform PaymentMethod into numerical columns
# This is another popular way to perform one-hot encoding

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['PaymentMethod']])
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(['PaymentMethod']))
encoded_df.head()
````

### Cell 6 — binary Yes/No → 1/0 mapping

````python
# Convert Yes/No values into 1/0 numbers
# Binary mapping is useful for machine learning models

df['Churn_bin'] = df['Churn'].map({'No': 0, 'Yes': 1})
df[['Churn', 'Churn_bin']].head()
````

### Cell 7 — encode multiple columns at once

````python
# Encode multiple categorical columns at the same time
# All selected text columns are converted into numerical features

multi = pd.get_dummies(df[['gender', 'Contract', 'PaymentMethod']], drop_first=True)
print('Shape:', multi.shape)
multi.head()
````
