## Impute numerical columns with median

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
# Step 7: Check & impute categorical columns with mode
for col in df.select_dtypes(include='object').columns:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
        print(f"Imputed '{col}' with mode: {df[col].mode()[0]}")

print("Categorical missing check done.")
```

# Standardize Inconsistent Categorical Entries

```python
df['gender'].value_counts()
```

```python
# To get unique values, you typically apply .unique() to a specific column (Series).
# For example, to see unique values in the 'gender' column:
df['gender'].unique()
```

To get unique values for all columns, we can loop through each column and apply the `.unique()` method.

```python
for column in df.columns:
    print(f"Unique values in column '{column}':\n{df[column].unique()}\n")
```

To get the frequency count of unique values for all columns, we can loop through each column and apply the `.value_counts()` method.

```python
for column in df.columns:
    print(f"Value counts for column '{column}':\n{df[column].value_counts()}\n")
```

From the `value_counts()` output, we can observe that columns like `MultipleLines`, `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, `TechSupport`, `StreamingTV`, and `StreamingMovies` contain 'No internet service' as a category. This can be standardized to 'No' to represent the absence of that particular service.

```python
# Replace 'No internet service' with 'No' for relevant columns
for col in ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']:
    if 'No internet service' in df[col].unique():
        df[col] = df[col].replace('No internet service', 'No')

# Also standardize 'No phone service' to 'No' in 'MultipleLines'
if 'No phone service' in df['MultipleLines'].unique():
    df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')

# Verify the changes by checking unique values again
print("Unique values after standardization:")
for col in ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']:
    print(f"'{col}': {df[col].unique()}")
```

To further standardize categorical features, it's often useful to convert all string entries to lowercase. This ensures that variations in casing (e.g., 'Yes' vs. 'yes') are treated as the same category.

```python
# Step 3: Strip leading/trailing whitespace
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

print("Whitespace stripped from all object columns.")
```

```python
# Convert all object type columns to lowercase
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.lower()

# Display unique values of a sample column to verify (e.g., 'gender')
print("Unique values in 'gender' after lowercasing:", df['gender'].unique())
```

### Unrealistic Values and Their Handling

1.  **`TotalCharges` Blank Strings:**
    *   **Detection:** Initially, `TotalCharges` was of object type, and `df[df['TotalCharges'] == ' ']` identified 11 rows where `TotalCharges` contained blank strings.
    *   **Handling:** These blank strings were converted to `NaN` during the `pd.to_numeric(df['TotalCharges'], errors='coerce')` step. A new indicator column, `TotalCharges_missing`, was created to mark these rows. Finally, the `NaN` values in `TotalCharges` were imputed with the median of the column.

2.  **`tenure = 0` with Non-Zero `TotalCharges`:**
    *   **Detection:** After imputation, it was observed that customers with a `tenure` of 0 months (meaning they just joined) still had a `TotalCharges` value equal to the median (1397.475). This is unrealistic, as a customer with 0 tenure should theoretically have 0 total charges.
    *   **Handling:** For the purpose of this analysis, these values were **left as is** following the median imputation step. It's noted as an inconsistency that might require further domain-specific handling (e.g., setting `TotalCharges` to 0 for these specific cases if `MonthlyCharges` was also 0, or if they represent initial setup fees). However, given the median imputation was already applied to handle missing values, and these were considered 'missing' (blanks) initially, this was the direct consequence of that imputation strategy.

3.  **Negative `MonthlyCharges`:**
    *   **Detection:** A check using `df[df['MonthlyCharges'] < 0]` was performed.
    *   **Handling:** No unrealistic negative values were found in the `MonthlyCharges` column.

4.  **Duplicate Rows:**
    *   **Detection:** Checks using `df.duplicated().sum()` and `df[df.duplicated(keep=False)]` were performed.
    *   **Handling:** No duplicate rows were found in the dataset, so no removal was necessary.
